# Series

```python
import pandas as pd
```

```python
x = [2,3,4,5,6]

var = pd.Series(x, index=['a','b','c','d','e'], dtype="float", name = "python")

print(var)
print(type(var))
print(var[2])
```

```python
dic =  {"name":['python','c','c++','java'],"rank":[1,2,3,4]}

var1 = pd.Series(dic)

print(var1)
```

```python
s = pd.Series(12,index=[1,2,3,4,5,6,7])

print(s)
print(type(s))
```

```python
s1 = pd.Series(12,index=[1,2,3,4,5,6,7])
s2 = pd.Series(12,index=[1,2,3,4])

print(s1+s2)
```

```python
varr_new = varr.groupby("Name")
```

# Join function in dataset

```python
var1 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]})
var2 = pd.DataFrame({"Enr_no" : [1,2,3,4,5],
                     "id_no" : [23,3,443,432,13]})

var1.join(var2)
```

```python
var2.join(var1)
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]})
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]})

var3.join(var4)
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]}, index=["a","b","c","d","e"])
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]},index=["a","b","c"])

var3.join(var4)
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]}, index=["a","b","c","d","e"])
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]},index=["a","b","c"])

var3.join(var4, how = "left")
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]}, index=["a","b","c","d","e"])
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]},index=["a","b","c"])

var3.join(var4, how = "right")
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]}, index=["a","b","c","d","e"])
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]},index=["a","b","c"])

var3.join(var4, how = "outer")
```

```python
var3 = pd.DataFrame({"name" : ["shyam","abhay","nishtha","darshan","harshil"],
                     "age" : [20,21,22,23,24]}, index=["a","b","c","d","e"])
var4 = pd.DataFrame({"Enr_no" : [1,2,3],
                     "id_no" : [23,3,443]},index=["a","b","c"])

var3.join(var4, how = "inner")
```
